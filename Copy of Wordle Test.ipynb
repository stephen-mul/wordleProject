## Wordle Game

```python
class WordlePlayer:
    def __init__(self, word, model, lives=6):
        self.original_word = word #set to the original string
        self.full_word = [ord(i) - 97 for i in word] #converts string to alphabet positions
        self.letters_guessed = set([]) #set of guessed letters
        self.letters_remainning = set(self.full_word) #initially full word
        self.lives_left = lives
        self.obscured_words_seen = [] #input to model
        self.letters_previously_guessed = [] #input to model
        self.guesses = [] #store model guesses
        self.correct_responses = []
        self.z = model
        return
        
        def encode_obscured_word(self):
            word = [i if i in self.letters_guessed else 26 for i in self.full_word] #for each letter in the word if letter has been guessed, set to that value, else 26
            obscured_word = np.zeros((len(word), 27), dtype=np.float32) #numpy array with width 27, height the length of the word
            for i, j in enumerate(word):
                obscured_word[i, j] = 1
            return(obscured_word)
                
        def encode_guess(self, guess):
            encoded_guess = np.zeros((len(guess), 27), dtype=np.float32)
            for i, j in enumerate(guess):
                encoded_guess[i, j] = 1
            return(encoded_guess)

        def encode_previous_guesses(self):
            # Create a 1 x 26 vector where 1s indicate that the letter was previously guessed
            guess = np.zeros(26, dtype=np.float32)
            for i in self.letters_guessed:
                guess[i] = 1
            return(guess)

        def encode_correct_responses(self):
            # To be used with cross_entropy_with_softmax, this vector must be normalised
            response = np.zeros(26, dtype=np.float32)
            for i in self.letters_remaining:
                response[i] = 1.0
            response /= response.sum()
            return(response)
            
        def store_guess_and_result(self, guess):
            # Record what the model saw as input: an obscured word and a list of previously-guessed letters
            self.obscured_words_seen.append(self.encode_obscured_word())
            self.letters_previously_guessed.append(self.encode_previous_guesses())

            # Record the word that the model guessed, and add that guess to the list of previous guesses
            self.guesses.append(guess)
            guess_as_list = list(guess)
            for letter in guess_as_list:
                if letter not in self.letters_guessed:
                    self.letters_guessed.add(letter)
            
            # Store the "correct responses"
            correct_responses = self.encode_correct_responses()
            self.correct_responses.append(correct_responses)

            # Determine an appropriate reward, and reduce # of lives left if appropriate
            for letter in guess_as_list:
                if letter in self.letters_remaining:
                    self.letters_remaining.remove(letter)

            for letter in guess_as_list:
                if self.correct_responses[-1][letter] < 0.00001:
                    self.lives_left -= 1
                    break

        def run(self):
            # Play a game until we run out of lives or letters
            while (self.lives_left > 0) and (len(self.letters_remaining) > 0):
                guess = np.argmax(np.)
                

    

        




    
```

```python
## Model

def create_model():
    # Input #
    input_layer = Input(shape=(5, 28, 1), dtype = np.float32)

    ############
    # Encoding #
    ############

    # Conv1 #
    conv1 = Conv2D(filters=14, kernel_size=(3,3), activation = 'relu', padding = 'same')(input_layer)
    maxPool1 = MaxPooling2D(pool_size = (2,2), padding = 'same')(conv1)

    # Conv2 #
    conv2 = Conv2D(filters = 7, kernel_size = (3,3), activation = 'relu', padding = 'same')(maxPool1)
    maxPool2 = MaxPooling2D(pool_size = (2,2), padding='same')(conv2)

    ############
    # Decoding #
    ############

    # DeConv1 #

    deConv1 = Conv2D(filters = 7, kernel_size = (3,3), activation = 'relu', padding = 'same')(maxPool2)
    upSamp1 = UpSampling2D((2,2))(deConv1)

    # DeConv2 #

    deConv2 = Conv2D(filters = 14, kernel_size = (3,3), activation = 'sigmoid', padding = 'same')(upSamp1)
    upSamp2 = UpSampling2D((2,2))(deConv2)

    # Cropping #
    crop = Cropping2D(cropping=((1, 0), (0,0)), data_format=None)(upSamp2)


    # Ouput #
    output_layer = Conv2D(filters = 1, kernel_size = (3,3), activation = 'sigmoid')(crop)

    #output_layer = Reshape(target_shape=(5,28,1))(conv2)

    model = Model(input_layer,output_layer)
    model.compile(loss='mean_squared_error', optimizer='Adamax', metrics=['accuracy'])
    return model
```

```python
# Get word function

def get_word():
    word = random.choice(five_letters)
    return word.upper()

def split_word(word):
    return list(word)
```

```python
# for gameplay

def play(word):
    word_completion = "_"*len(word)
    guessed = False
    #guessed_letters = []
    guessed_words = []
    correct_letters = []
    correct_position = []
    tries = 6
    print("Let's play Wordle !")
    print(tries)
    print(word_completion)
    print("\n")
    while not guessed and tries > 0:
        guess = input("Please guess a word: ").upper()
        if len(guess) == len(word) and guess.isalpha():
            word_letters = split_word(word)
            guess_letters = split_word(guess)
            if guess in guessed_words:
                print("You already guessed the word", guess)
            elif guess != word:
                print(guess, "is not the word.")
                tries -= 1
                guessed_words.append(guess)
                count = 0
                for letter in guess_letters:
                    if letter in word_letters:
                        if guess_letters.index(letter) == word_letters.index(letter):
                            print(letter, "is in the right position.")
                            word_as_list = list(word_completion)
                            word_as_list[count] = letter
                            word_completion = "".join(word_as_list)
                            if "_" not in word_completion:
                                guessed = True
                        else:
                            print(letter, "is in the word, but not in the right position.")
                    count += 1
                        
            else:
                guessed = True
                word_completion = word
        else:
            print("Not a valid guess.")
            
        print(tries, "remaining tries.")
        print(word_completion)
        print("\n")
    if guessed:
        print("Congrats, you guessed the word! You win!")
    else:
        print("Sorry, you ran out of tries. The word was " + word + ". Maybe next time.")
```

```python
def main ():
    word = get_word()
    play(word)
    while input("Play Again? (Y/N)").upper() == "Y":
        word = get_word()
        play(word)
        
if __name__ == "__main__":
    main()
```
